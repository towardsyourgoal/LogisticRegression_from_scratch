```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
import category_encoders as ce
```

```python
%load_ext autoreload
%autoreload 2
```

```python
from LogisticRegressor import LogisticRegressor as LogReg
```

```python
car_data=pd.read_csv('/File_path/car_evaluation.csv')
car_data.columns=['buying_price','maint_price','n_doors','n_persons','lug_boot','safety','decision']
car_data['decision']=car_data['decision'].replace({'good':'acc','vgood':'acc'})
col_names=car_data.columns.drop(['decision'])

#OneHotEncoding for predictors
ce_OHE = ce.OneHotEncoder(cols=col_names) 
ce_OHE
car_data_encoded = ce_OHE.fit_transform(car_data)

#label encoding for target variable
mapping={'acc':0,'unacc':1}
car_data_encoded['decision_encoded']=car_data_encoded['decision'].map(mapping)

X=car_data_encoded.drop(['decision','decision_encoded'],axis=1)
y=car_data_encoded['decision_encoded']

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
log_reg_model=LogReg()
```

```python
log_reg_model.fit(X_train,y_train)
```

```python
y_pred=log_reg_model.predict(X_test)
```

```python
y_score=log_reg_model.predict_probs(X_test)
```

```python
mse=log_reg_model.mse(y_test,y_pred)
acc=log_reg_model.accuracy_score(y_test,y_pred)
prec=log_reg_model.precision_score(y_test,y_pred)
recall=log_reg_model.recall_score(y_test,y_pred)
f1score=log_reg_model.f1_score(y_test,y_pred)
fpr=log_reg_model.fpr(y_test,y_pred)
specificity=log_reg_model.specificity(y_test,y_pred)
roc_auc_score=log_reg_model.calculate_roc_auc_score(y_test,y_pred,y_score)

print("MSE is: ",mse)
print("accuracy is: ",acc)
print("precision is: ",prec)
print("recall is: ",recall)
print("f1 score is: ",f1score)
print("FPR is: ",fpr)
print("Specificity is: ",specificity)
print("ROC AUC Score is: ",roc_auc_score)
```
